```python
# Remove duplicates from gender_submissiondata
gender_submissiondata.drop_duplicates(inplace=True)
# Remove duplicates from train_data
train_data.drop_duplicates(inplace=True)
# Remove duplicates from test_data
test_data.drop_duplicates(inplace=True)
```

```python
non_numeric_cols = test_data.select_dtypes(include=["object"]).columns.tolist()
test_data_numeric = test_data.select_dtypes(include=["int64", "float64"])
test_data_numeric.fillna(test_data_numeric.mean(), inplace=True)
test_data
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

train_data = remove_outliers(train_data, 'Age')
train_data = remove_outliers(train_data, 'Fare')
sns.pairplot(train_data[['Age', 'Fare', 'Survived']],hue='Survived')
plt.show()
```

```python
sns.pairplot(train_data[['Age', 'Fare', 'Survived']],hue='Age')
plt.show()
```

```python
sns.pairplot(train_data[['Age', 'Fare', 'Survived']],hue='Fare',palette='Greens')
plt.show()
```

```python
# Exploratory Data Analysis (EDA) for train_data
# Summary statistics
print("Train Data Summary Statistics:")
print(train_data.describe())

# Correlation matrix
correlation_matrix_train = train_data[['Age', 'Fare', 'Survived']].corr()
print("\nTrain Data Correlation Matrix:")
print(correlation_matrix_train)

# Countplot for 'Survived'
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=train_data,color='pink',edgecolor='darkred')
plt.title('Survival Count in Train Data')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Age', data=train_data,color='brown')
plt.title('Distribution of Age by Survival Status in Train Data')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Fare', data=train_data,color='lightblue')
plt.title('Distribution of Fare by Survival Status in Train Data')
plt.show()
```
